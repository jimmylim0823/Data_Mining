```python
import pandas as pd
dataset_total = pd.read_csv('/content/drive/Shareddrives/데마-돼지귀엽조/프로젝트/csv 정리 파일/total_revised.csv', encoding='cp949')
#dataset_total = pd.read_csv('total_revised.csv', encoding='cp949')
dataset_total.head()
```

```python
df_total = pd.DataFrame()
df_total['content'] = dataset_total['content_of_accdient']
df_total['answer'] = dataset_total['answer_content']
df_total
```

```python
df_total.replace("", float("NaN"), inplace=True)
df_total.dropna(inplace=True)
print('total: ', len(df_total))
```

```python
def check_result(d):
    res = float('nan')
    if '종료' in d:
        res = 1
    elif '종결' in d:
        res = 1
    elif '처리불가' in d:
        res = 0
    elif '양해' in d:
        res = 0
    else:
        float('NaN')
    return res
```

```python
df_total['result'] = df_total['answer'].apply(check_result)
print('closed :', len(df_total[df_total['result']==1]))
print('rejected :', len(df_total[df_total['result']==0]))
```

```python
df_total.replace("", float("NaN"), inplace=True)
df_total.dropna(inplace=True)
print('total: ', len(df_total))
```

```python
df_total.drop('answer', axis=1, inplace=True)
```

```python
def check_six_ws(d):
    six_ws = ["언제", "누가", "어디서", "무엇을", "어떻게", "왜"]
    res = 1
    for word in six_ws:
        if word not in d:
            res = 0
            break
        else:
            continue
    return res
```

```python
df_total['six_ws'] = df_total['content'].apply(check_six_ws)
df_total.head()
```

```python
import re

def clean_text(d):
    pattern = '[^ㄱ-힗\s]'
    text = re.sub(pattern, '', d)
    return text
```

```python
df_total['content'] = df_total['content'].apply(clean_text)
df_total.head()
```

```python
stop_word = pd.read_csv('/content/drive/Shareddrives/데마-돼지귀엽조/프로젝트/불용어.txt', names=['stop'], header=None)
stop_words = stop_word['stop'].tolist()
stop_word
```

```python
def clean_stopword(d, stop_words):
    stop_words = stop_words + ["언제", "누가", "어디서", "무엇을", "어떻게", "왜", "년", "월", "일"]
    word_list = d.split(' ')
    result = []

    for word in word_list:
        temp = True

        for stop in stop_words:
            if stop in word:
                word_split = word.partition(stop)
                for w in word_split:
                    if w == '':
                        continue
                    elif w in stop_words:
                        continue
                    else:
                        result.append(w)

                temp = False
                break
            else:
                continue

        if temp == True:
            result.append(word)
        else:
            continue
    
    text = ' '.join(result)
    return text
```

```python
df_total['content'] = df_total['content'].apply(clean_stopword, stop_words=stop_words)
df_total.head()
```

```python
from konlpy.tag import Okt
okt = Okt()

def preprocessing(x):
    text = []
    tokens = okt.pos(x)
    for token in tokens :
        if token[1] == 'Adjective' or token[1]=='Adverb' or token[1] == 'Determiner' or token[1] == 'Noun' or token[1] == 'Verb' or 'Unknown':
            text.append(token[0])
    return text
```

```python
df_total['content'] = df_total['content'].apply(preprocessing)
df_total.head()
```

```python
pos_pol_word = pd.read_csv('/content/drive/Shareddrives/데마-돼지귀엽조/프로젝트/pos_pol_word.txt', names=['pos'], header=None)
neg_pol_word = pd.read_csv('/content/drive/Shareddrives/데마-돼지귀엽조/프로젝트/neg_pol_word.txt', names=['neg'], header=None)
neg_pol_word.head()
```

```python
pos_pol_word[0] = pos_pol_word['pos']
pos_pol_word[1] = 1
neg_pol_word[0] = neg_pol_word['neg']
neg_pol_word[1] = -1
sent_dic = pd.merge(pos_pol_word, neg_pol_word, 'outer')[[0,1]]
sent_dic
```

```python
def get_score(x):
    cnt = 0
    for word in list(set(x)):
        target = sent_dic[sent_dic[0]==word]
        if len(target)==1:
            cnt += float(target[1])
    return cnt
```

```python
df_total['score'] = df_total['content'].apply(get_score)
df_total
```

```python
import numpy as np
df_total['ratio'] = df_total['score'] / (np.log10(len(df_total['content'])) + 1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(df_total['ratio']).set_title("Sentiment Ratio")
plt.show()
```

```python
def get_count(x):
    cnt = 0
    for word in list(set(x)):
        target = sent_dic[sent_dic[0]==word]
        if len(target)==1:
            cnt += 1
    return cnt
```

```python
df_total['count'] = df_total['content'].apply(get_count)
df_total
```

```python
print('어휘 매칭 성공: ', len(df_total[df_total['count']!=0]))
print('어휘 매칭 실패: ', len(df_total[df_total['count']==0]))
```

```python
df_total['count'].replace(0, float("NaN"), inplace=True)
df_total.dropna(inplace=True)
print('total: ', len(df_total))
```
